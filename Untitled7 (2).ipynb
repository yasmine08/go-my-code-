```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
    
```

```python
plot_correlation_map( df )
```

```python
x=df['Age'].values[:,np.newaxis]
y=df['Pclass'].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40)
model=LinearRegression()
model.fit(x_train,y_train)
predicted=model.predict(x_test)
print('MSE',mean_squared_error(y_test,predicted))
print('R squared',metrics.r2_score(y_test,predicted))
```

```python
#features extraction
x=df[['Age','Pclass','Sex']]
y=df['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40)
logreg=LogisticRegression() #build the logistic model
logreg.fit(x_train,y_train) #fitting trainnning data
y_pred=logreg.predict(x_test) #testing model's performance 
print("Accuracy={:.2f}".format(logreg.score(x_test,y_test)))
```

```python
sns.regplot(x='Age',y='Survived',data=df,logistic=True)
```

```python

print(classification_report(y_test,y_pred))
```

```python
confusion_matrix=pd.crosstab(y_test,y_pred,rownames=['Age'],colnames=['Survived'])
print(confusion_matrix)
```
